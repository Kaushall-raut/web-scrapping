```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

final_data = []

for j in range(1, 19):
    URL = f"https://www.ambitionbox.com/list-of-companies?page={j}"
    headers = {'User-Agent': UserAgent().random}
    
    response = requests.get(URL, headers=headers).text
    soup = BeautifulSoup(response, 'lxml')

    company = soup.find_all('div', class_="companyCardWrapper")

    for i in company:
        name = i.find('h2').text.strip() if i.find('h2') else None

        rating = (
            i.find('div', class_='rating_text').text.strip()
            if i.find('div', class_='rating_text') else None
        )

        action_counts = i.find_all('span', class_="companyCardWrapper__ActionCount")

        reviews = action_counts[0].text.strip() if len(action_counts) > 0 else None
        total_jobs = action_counts[-1].text.strip() if len(action_counts) > 0 else None

        final_data.append([name, rating, reviews, total_jobs])

# Create DataFrame ONCE
final_data = pd.DataFrame(
    final_data,
    columns=['name', 'rating', 'reviews', 'total_jobs']
)

final_data.shape
```
